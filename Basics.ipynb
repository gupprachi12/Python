```python
#Sum of Even Numbers: Write a program that calculates the sum of all even numbers between 1 and a given number n. Get n as input from the user.
#user_number = int(input('write a number'))
user_number = 7
even_number = 0
for number in range(1, user_number, 1):
  #print (number)
  if number % 2 == 0:
    even_number = even_number + number
    print(even_number)


```

```python
#Factorial Calculation: Write a program that calculates the factorial of a given number n. Get n as input from the user. (Factorial of n is the product of all positive integers less than or equal to n)
number_end = 4
product = 1
pos = 0
previous_product = []
for number in range(1, number_end+1):
  product = number * product
  previous_product.append(product)
  print(number ,'*', previous_product[pos-1], '=', product)
  pos = pos +1
```

```python
#Fibonacci Sequence: Write a program that prints the first n numbers of the Fibonacci sequence.  Get n as input from the user. (The Fibonacci sequence starts with 0 and 1, and each subsequent number is the sum of the two preceding ones.)
user_number = 6
Fibonacci_sequence = [0,1]
pos = 0
while user_number >= pos:  #range (0, user_number):
  if (Fibonacci_sequence[-1] + Fibonacci_sequence[-2] )== user_number or (Fibonacci_sequence[-1] + Fibonacci_sequence[-2]) > user_number:
    print('codeend')
    break
  Fibonacci_sequence.append(Fibonacci_sequence[-1] + Fibonacci_sequence[-2])
  print(Fibonacci_sequence)
  pos = pos + 1

print('Final output:' , Fibonacci_sequence)
```

```python
#List Comprehension Practice: Create a list of the squares of all odd numbers between 1 and 50 using a list comprehension.
odd_numbers = []
for number in range(0,50):
  if number % 2 != 0:
    odd_numbers.append(number*number)
print(odd_numbers)
```
